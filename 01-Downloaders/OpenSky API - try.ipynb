```python
import pickle
os.chdir("../../FlightPathAnalysis/01-Downloaders/")
def query_api_till_result(start_window, end_window, timeout = 15, attempts = 0):
    x = None
    while type(x) != list:
        attempts += 1
        print(attempts)
        try:
            x = api.get_flights_from_interval(begin = start_window, end = end_window, timeout = timeout)
        except:
            x = query_api_till_result(start_window, end_window, timeout, attempts)
        
        time.sleep(8)
    
    return x

for i, (start_window, end_window) in enumerate(time_windows):
    print(i, start_window, end_window, datetime.datetime.now())
    x = query_api_till_result(int(start_window), int(end_window), timeout = 15)
    pickle.dump(x, open("../data/" + str(int(start_window)) + "_" + str(int(end_window)) + ".pkl", "wb"))
```
